```python
#clean up dataset
df_c.drop(['lsoa_code', 'value', 'minor_category'], 
          axis=1, inplace=True)

#rename colomns
df_c.rename(columns={'major_category':'kejahatan_besar', 
                     'borough':'wilayah', 'month':'bulan'}, inplace=True)

df_c.columns = list(map(str, df_c.columns))

# df_c.set_index('wilayah', inplace=True)

#add total column
df_c['Total'] = df_c.sum(axis=1)

print('data dimension', df_c.shape)
```

```python

# pengambilan data 5 wilayah dengan 5 tahun
top_5_wilayah = df_c.groupby(['year', 'wilayah'])['Total'].sum().reset_index()
top_5_wilayah = top_5_wilayah[(top_5_wilayah['year'] > 2008) & (top_5_wilayah['year'] <2014)]

top_5_wilayah = top_5_wilayah.loc[(top_5_wilayah['wilayah'] == 'Wandsworth') | (top_5_wilayah['wilayah'] == 'Tower Hamlets') | (top_5_wilayah['wilayah'] == 'Bexley') | (top_5_wilayah['wilayah'] == 'Barking and Dagenham') | (top_5_wilayah['wilayah'] == 'Barnet')]
top_5_wilayah
```

```python
xs = top_5_wilayah['wilayah'].unique().tolist()
lists= []
lists_final = []
lists_final.append(xs)

for x in xs:
    lists.append(top_5_wilayah[top_5_wilayah['wilayah']==x]['Total'].values.tolist())
lists2 = list(map(list, zip(*lists)))

for x in lists2:
    lists_final.append(x)
dict = {'wilayah': lists_final[0], '2019':  lists_final[1], '2020':  lists_final[2], '2021':  lists_final[3], '2022':  lists_final[4], '2023' :  lists_final[5]} 
df_top5_wilayah_5tahun = pd.DataFrame(dict)
    
df_top5_wilayah_5tahun 
```

```python
df_top5_wilayah_5tahun.transpose()
```

```python
years = [2019,2020,2021,2022,2023]
years
```

```python
list_top5_transpose = df_top5_wilayah_5tahun.transpose().values.tolist()[1:]
dict = {'Barking and Dagenham': list_top5_transpose[4], 'Barnet':  list_top5_transpose[3], 'Bexley':  list_top5_transpose[2], 'Tower Hamlets':  list_top5_transpose[1], 'Wandsworth':  list_top5_transpose[0] } 
df_top5_transpose = pd.DataFrame(dict, index=[years])
df_top5_transpose
```

```python
#visualisasi area
df_top5_transpose.index = years# let's change the index values of df_top5 to type integer for plotting

df_top5_transpose.plot(kind='area', 
             stacked=True,
             figsize=(20, 10), # pass a tuple (x, y) size
             )


plt.xlim([2019,2023])

plt.show()
```
